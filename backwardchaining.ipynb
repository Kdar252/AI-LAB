```python
class KnowledgeBase:
    def __init__(self):
        self.facts = set()
        self.rules = []

    def add_fact(self, fact):
        self.facts.add(fact)

    def add_rule(self, rule):
        self.rules.append(rule)

    def backward_chaining(self, goal):
        return self.backward_chaining_recursive(goal, set())

    def backward_chaining_recursive(self, goal, visited):
        if goal in self.facts:
            return True

        for rule in self.rules:
            if goal == rule.conclusion:
                if all(self.backward_chaining_recursive(premise, visited) for premise in rule.premises):
                    return True

        return False


class Rule:
    def __init__(self, premises, conclusion):
        self.premises = premises
        self.conclusion = conclusion


# Example usage:
kb = KnowledgeBase()

# Add facts to the knowledge base
kb.add_fact('A')
kb.add_fact('B')
kb.add_fact('C')

# Define rules
rule1 = Rule(['A', 'B'], 'D')
rule2 = Rule(['D', 'C'], 'E')

# Add rules to the knowledge base
kb.add_rule(rule1)
kb.add_rule(rule2)

# Use backward chaining to check if 'E' can be inferred
goal = 'E'
result = kb.backward_chaining(goal)

if result:
    print(f"The goal '{goal}' can be inferred.")
else:
    print(f"The goal '{goal}' cannot be inferred.")
```
